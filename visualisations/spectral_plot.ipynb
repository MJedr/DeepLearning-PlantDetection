```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_pickle(r'')
```

```python
dataset.columns = ['x', 'y', 'ekstrakcja', 'klasa', 'indeks']
```

```python
x_col_name = 'ekstrakcja'
y_col_name = 'klasa'
```

```python
X = dataset[x_col_name].apply(lambda x: np.array(x)).values
X = np.concatenate(X).reshape(X.shape[0], X[1].shape[0], 1)
X = (X - np.min(X))/(np.max(X) - np.min(X))
y = dataset[y_col_name].values
```

```python
classes = dataset.klasa.unique()
```

```python
classes_eng = ['Shadows', 'Roads', 'Coniferous trees', 
              'Dediceous trees', 'Filipendula ulmaria',
              'Molinia caeruela', 'Cultivated areas',
              'Grasslands', 'Buildings']
```

```python
class0 = np.mean(X[np.where(y == classes[0])[0]], axis=0)
class1 = np.mean(X[np.where(y == classes[1])[0]], axis=0)
class2 = np.mean(X[np.where(y == classes[2])[0]], axis=0)
class3 = np.mean(X[np.where(y == classes[3])[0]], axis=0)
class4 = np.mean(X[np.where(y == classes[4])[0]], axis=0)
class5 = np.mean(X[np.where(y == classes[5])[0]], axis=0)
class6 = np.mean(X[np.where(y == classes[6])[0]], axis=0)
class7 = np.mean(X[np.where(y == classes[7])[0]], axis=0)
class8 = np.mean(X[np.where(y == classes[8])[0]], axis=0)
```

```python
class0STD = np.std(X[np.where(y == classes[0])[0]], axis=0)
class1STD = np.std(X[np.where(y == classes[1])[0]], axis=0)
class2STD = np.std(X[np.where(y == classes[2])[0]], axis=0)
class3STD = np.std(X[np.where(y == classes[3])[0]], axis=0)
class4STD = np.std(X[np.where(y == classes[4])[0]], axis=0)
class5STD = np.std(X[np.where(y == classes[5])[0]], axis=0)
class6STD = np.std(X[np.where(y == classes[6])[0]], axis=0)
class7STD = np.std(X[np.where(y == classes[7])[0]], axis=0)
class8STD = np.std(X[np.where(y == classes[8])[0]], axis=0)
```

```python
class0minusSTD = class0 - class0STD
class1minusSTD  = class1 - class0STD
class2minusSTD = class2 - class0STD
class3minusSTD  = class3 - class0STD
class4minusSTD = class4 - class0STD
class5minusSTD  = class5 - class0STD
class6minusSTD  = class6 - class0STD
class7minusSTD  = class7 - class0STD
class8minusSTD  = class8 - class0STD
```

```python
class0plusSTD = class0 + class0STD
class1plusSTD  = class1 + class0STD
class2plusSTD = class2 + class0STD
class3plusSTD  = class3 + class0STD
class4plusSTD = class4 + class0STD
class5plusSTD  = class5 + class0STD
class6plusSTD  = class6 + class0STD
class7plusSTD  = class7 + class0STD
class8plusSTD  = class8 + class0STD
```

```python
fig, ax = plt.subplots(3, 3, figsize=(20,10))
ax[0, 0].plot(class0)
ax[0, 0].fill_between(np.arange(class0.shape[0]), class0minusSTD.flatten(), class0plusSTD.flatten(), color='skyblue')
ax[0, 0].set_title(classes_eng[0])
ax[0, 0].grid(axis='y')

ax[1, 0].plot(class1)
ax[1, 0].fill_between(np.arange(class1.shape[0]), class1minusSTD.flatten(), class1plusSTD.flatten(), color='skyblue')
ax[1, 0].set_title(classes_eng[1])
ax[1, 0].grid(axis='y')

ax[2, 0].plot(class2)
ax[2, 0].fill_between(np.arange(class2.shape[0]), class2minusSTD.flatten(), class2plusSTD.flatten(), color='skyblue')
ax[2, 0].set_title(classes_eng[2])
ax[2, 0].grid(axis='y')

ax[0, 1].plot(class3)
ax[0, 1].fill_between(np.arange(class3.shape[0]), class3minusSTD.flatten(), class3plusSTD.flatten(), color='skyblue')
ax[0, 1].set_title(classes_eng[3])
ax[0, 1].grid(axis='y')

ax[1, 1].plot(class4)
ax[1, 1].fill_between(np.arange(class4.shape[0]), class4minusSTD.flatten(), class4plusSTD.flatten(), color='skyblue')
ax[1, 1].set_title(classes_eng[4])
ax[1, 1].grid(axis='y')

ax[2, 1].plot(class5)
ax[2, 1].fill_between(np.arange(class5.shape[0]), class5minusSTD.flatten(), class5plusSTD.flatten(), color='skyblue')
ax[2, 1].set_title(classes_eng[5])
ax[2, 1].grid(axis='y')

ax[0, 2].plot(class6)
ax[0, 2].fill_between(np.arange(class6.shape[0]), class6minusSTD.flatten(), class6plusSTD.flatten(), color='skyblue')
ax[0, 2].set_title(classes_eng[6])
ax[0, 2].grid(axis='y')

ax[1, 2].plot(class7)
ax[1, 2].fill_between(np.arange(class7.shape[0]), class7minusSTD.flatten(), class7plusSTD.flatten(), color='skyblue')
ax[1, 2].set_title(classes_eng[7])
ax[1, 2].grid(axis='y')

ax[2, 2].plot(class8)
ax[2, 2].fill_between(np.arange(class8.shape[0]), class8minusSTD.flatten(), class8plusSTD.flatten(), color='skyblue')
ax[2, 2].set_title(classes_eng[8])
ax[2, 2].grid(axis='y')

fig.tight_layout() 
plt.savefig('wykres_odbic_ts.png')
```
